## Data

We will get stock information for the following banks:
*  Bank of America
* CitiGroup
* Goldman Sachs
* JPMorgan Chase
* Morgan Stanley
* Wells Fargo


```python
start = datetime.datetime(2006, 1, 1)
end = datetime.datetime(2016, 1, 1)
```

```python
#To display all the columns
pd.set_option('display.max_columns', 50)
```

```python
BAC = client.get_dataframe("BAC",startDate=start,endDate=end,frequency='daily')
#Dropping these columns because we won't need them
BAC.drop(['adjClose','adjHigh','adjLow','adjOpen','adjVolume','divCash','splitFactor'],axis=1,inplace=True)
```

```python
c=client.get_dataframe("C",startDate=start,endDate=end,frequency='daily')
c.drop(['adjClose','adjHigh','adjLow','adjOpen','adjVolume','divCash','splitFactor'],axis=1,inplace=True)
```

```python
gs=client.get_dataframe("GS",startDate=start,endDate=end,frequency='daily')
gs.drop(['adjClose','adjHigh','adjLow','adjOpen','adjVolume','divCash','splitFactor'],axis=1,inplace=True)
```

```python
jpm=client.get_dataframe("JPM",startDate=start,endDate=end,frequency='daily')
jpm.drop(['adjClose','adjHigh','adjLow','adjOpen','adjVolume','divCash','splitFactor'],axis=1,inplace=True)
```

```python
ms=client.get_dataframe("MS",startDate=start,endDate=end,frequency='daily')
ms.drop(['adjClose','adjHigh','adjLow','adjOpen','adjVolume','divCash','splitFactor'],axis=1,inplace=True)
```

```python
wfc=client.get_dataframe("WFC",startDate=start,endDate=end,frequency='daily')
wfc.drop(['adjClose','adjHigh','adjLow','adjOpen','adjVolume','divCash','splitFactor'],axis=1,inplace=True)
```

### Create a list of the ticker symbols (as strings) in alphabetical order because we will be needing it later.

```python
tickers=['BAC','C','GS','JPM','MS','WFC']
```

#### I will use pd.concat to concatenate the bank dataframes together to a single data frame called bank_stocks. I will set the keys argument equal to the tickers list.

```python
bank_stocks=pd.concat([BAC,c,gs,jpm,ms,wfc],axis=1,keys=tickers)
```

#### I will set the column name levels to Bank Ticker and Stock Info.

```python
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
```

#### I will check the head of the bank_stocks dataframe to get a feeling of what Data that I'm going to handle. 

```python
bank_stocks.head(2)
```

# EDA

#### What is the max Close price for each bank's stock throughout the time period?

```python
bank_stocks.xs(key='close',axis=1,level='Stock Info').max()
```

#### I will create a new empty DataFrame called returns. This dataframe will contain the returns for each bank's stock.

```python
returns=pd.DataFrame()
```

#### We can use pandas pct_change() method on the Close column to create a column representing this return value. I will create a for loop that goes and for each Bank Stock Ticker creates this returns column and set's it as a column in the returns DataFrame. Here we will use the tickers list


```python
for tick in tickers:
   returns[tick+' Return'] = bank_stocks[tick]['close'].pct_change()
returns.head()
```

#### Here I will create a pairplot using seaborn of the returns dataframe. As you can see the Citi group bank has almost 0 returns and we will see why

```python
#(1:) because the first day was Nan
sns.pairplot(data=returns[1:])
```

#### Using this returns DataFrame, we will see what dates each bank stock had the best and worst single day returns. You should notice that 3 banks share the same day, and that day was the market crash 


```python
returns.idxmin()
```

```python
returns.idxmax()
```

#### We will take a look at the standard deviation of the returns, after checking the standard deviation we can see that Citi Group is the riskiest stock since it has a high standard deviation.

```python
returns.std()
```

#### I will create a distplot using seaborn of the 2008 returns for CitiGroup to see how they did during the crash

```python
sns.distplot(returns.ix['2008-01-01':'2008-12-31']['C Return'],bins=100,color='red',hist_kws=dict(alpha=0.5))
```
